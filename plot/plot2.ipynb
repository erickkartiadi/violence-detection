```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
import textwrap


# Sample data
models = ['ConvLSTM2D', 'Conv3D + FD + MB', 'ConvLSTM2D + FD', 'ConvLSTM2D + FD + MB', 'ConvLSTM2D + FD + MB (FT)']
accuracy = [77.25, 77.5, 81.25, 84.5, 86.00]
colors = ['darkgray', 'darkgray', 'darkgray', 'limegreen', 'limegreen']  # The last two bars will be orange
wrapped_labels = [textwrap.fill(label, width=10) for label in models]

# Create the bar plot
plt.bar(wrapped_labels, accuracy, color=colors)
plt.ylabel('Accuracy (%)')
plt.xlabel('Model')
plt.title('')

# Adjust the y-axis limits to add more height
plt.ylim(0, 100)  # Adjust the upper limit as needed

# Create custom legend entries without colors
legend_text = mpatches.Patch(color='none', label='FD: Frame Difference\nMB: Motion Blur\nFT: Fine-tuned')

# Display the legend with no color marker space
plt.legend(
    handles=[legend_text],
    loc='lower right', 
    bbox_to_anchor=(1, 1),
    handlelength=0,  # Remove space for the color marker
    handletextpad=0,  # Reduce space between the handle and the text to zero
    # frameon=False  # Optional: Removes the box around the legend
)


# Display the plot
plt.tight_layout()
plt.show()
```
